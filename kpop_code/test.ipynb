```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) 파일에서 프레임별 유사도 파싱
file_path = 'results.txt'
frames = []
scores = []
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('Frame'):
            parts = line.split(':', 1)
            idx_part, score_part = parts[0], parts[1]
            try:
                idx = int(idx_part.split()[1])
                score = float(score_part)
            except ValueError:
                continue
            frames.append(idx)
            scores.append(score)

# 2) DataFrame 생성
df = pd.DataFrame({'frame': frames, 'score': scores})

# 3) 30프레임씩 묶어 초 단위로 평균 유사도 계산
fps = 30
df['second'] = df['frame'] // fps
sec_df = df.groupby('second', as_index=False)['score']\
           .mean()\
           .rename(columns={'score': 'mean_score'})

# 4) 초별 평균 유사도 콘솔 출력
print("=== 초별 평균 유사도 (30프레임 단위) ===")
print(sec_df.to_string(index=False))

# 5) 초별 평균 유사도 그래프
plt.figure(figsize=(10, 4))
plt.plot(sec_df['second'], sec_df['mean_score'], marker='o')
plt.title('초별 평균 유사도 (30프레임 간격)')
plt.xlabel('Second')
plt.ylabel('Average Similarity')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1) 프레임별 각도 차이 불러오기 (라디안)
angle_diffs = np.load('angle_diffs.npy')   # shape (T, 8)
fps         = 30
T, M        = angle_diffs.shape

# 2) 프레임 → 초 매핑 후 초별 평균 (라디안)
seconds      = np.arange(T) // fps         # 각 프레임이 속한 초 (정수)
df_frame     = pd.DataFrame(angle_diffs, columns=[
    'left_sh-el-wr','right_sh-el-wr',
    'left_el-wr-idx','right_el-wr-idx',
    'left_hip-kn-an','right_hip-kn-an',
    'left_kn-an-ft','right_kn-an-ft'
])
df_frame['second'] = seconds
df_sec_rad = df_frame.groupby('second').mean()

# 3) 라디안 → 도(°) 변환
df_sec_deg = np.rad2deg(df_sec_rad)
df_sec_deg['second'] = df_sec_deg.index

# 4) 결과 파일에서 프레임별 유사도 파싱
frames = []
scores = []
records = []
with open('results.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('Frame'):
            parts = line.split(':', 1)
            try:
                frame_idx = int(parts[0].split()[1])
                score_val = float(parts[1])
            except:
                continue
            records.append((frame_idx, score_val))

# 이제 records 에는 (frame, score) 튜플만 들어 있으므로 길이가 같습니다.
df_score = pd.DataFrame(records, columns=['frame', 'score'])
df_score['second'] = df_score['frame'] // 30
sec_df = df_score.groupby('second', as_index=False)['score'].mean()

# # 5) 초별 평균 유사도 바 차트
# plt.figure(figsize=(12, 5))
# plt.bar(sec_df['second'], sec_df['score'], width=0.8)
# plt.xlabel('초 (s)')
# plt.ylabel('평균 유사도')
# plt.title('초별 평균 유사도 (30프레임 묶음)')
# plt.xticks(sec_df['second'])
# plt.grid(axis='y', linestyle='--', alpha=0.5)
# plt.tight_layout()
# plt.show()

# 6) 관절별 개별 그래프 (° 단위) ===
JOINT_LABELS = [
    'left_sh-el-wr','right_sh-el-wr',
    'left_el-wr-idx','right_el-wr-idx',
    'left_hip-kn-an','right_hip-kn-an',
    'left_kn-an-ft','right_kn-an-ft'
]

# xtick 레이블: sec0(0–29), sec1(30–59), …
secs = df_sec_deg['second'].tolist()
xtick_labels = [f"sec{sec}\n({sec*30}-{sec*30+29})" for sec in secs]

for label in JOINT_LABELS:
    plt.figure(figsize=(8, 3))
    plt.plot(df_sec_deg['second'], df_sec_deg[label],
             marker='o', markersize=3, linestyle='-')
    plt.title(f"{label} 각도 차이 추이")
    plt.xticks(secs, xtick_labels, rotation=45, ha='right')
    plt.xlabel("초 (s) 및 프레임 구간")
    plt.ylabel("평균 각도 차이 (°)")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```
